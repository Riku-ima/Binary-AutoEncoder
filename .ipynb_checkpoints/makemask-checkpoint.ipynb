```python
import cv2
import matplotlib.pyplot as plt
import csv
import numpy as np
import glob
import os
%matplotlib inline
```

```python
"""
まず拡張前の画像でマスクを作成
"""

point_List=[]
imgs_path='./Data/images/KAWASAKIvsCHIBA/'
csvs_path='./Data/anno/KAWASAKIvsCHIBA/*'
csvs=sorted(glob.glob(csvs_path))

for i,csv_path in enumerate(csvs):
    #csvのある画像だけmask
    file_name=os.path.splitext(os.path.basename(csv_path))[0]
    img_path=imgs_path+file_name+'.jpg'
    img=cv2.imread(img_path)
    print(csv_path)
    with open(csv_path) as f:
        reader=csv.reader(f)
        for row in reader:
            a=(int(row[0]),int(row[1]))
            point_List.append(a)
    img[:,:,:]=(0,0,0)
    cv2.fillPoly(img,[np.array(point_List)],(255,255,255))
    save_mask_path=os.path.join('./Data/mask_imges/KAWASAKIvsCHIBA',os.path.basename(img_path))
    cv2.imwrite(save_mask_path,img)
    point_List=[]
```

```python
#組み合わせ検証

#img_path=sorted(glob.glob('./Data/images/KAWASAKIvsCHIBA/*'))
mask_path=sorted(glob.glob('./Data/mask_imges/KAWASAKIvsCHABA/*'))
print(mask_path)


for mask in (mask_path):
    file_name=os.path.splitext(os.path.basename(mask))[0]
    img_path='./Data/images/KAWASAKIvsCHIBA/'+file_name+'.jpg'
    Img=cv2.imread(img_path)
    Mask=cv2.imread(mask)

    fig=plt.figure()
    X=2
    Y=2
    ax1=fig.add_subplot(X,Y,1)
    #タイトルの設定
    ax1.set_title("oriimg",fontsize=15)
    plt.imshow(cv2.cvtColor(Img,cv2.COLOR_BGR2RGB))

    ax2=fig.add_subplot(X,Y,2)
    #タイトルの設定
    ax2.set_title("maskimg",fontsize=15)
    plt.imshow(Mask)

    
```
